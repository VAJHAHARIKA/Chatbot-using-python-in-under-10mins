**OPTIONAL THINGS TO KNOW**

NLTK: Stands for Natural Language Tool Kit.
* It is a pre-defined library that works with human lauguage. 
* It consists of libraries that process text like tokenisation, parsing, classification, stemming, lemmatisation, tagging etc.

**Tokenisation:** It is the division of sentence. 
Ex: Hello Damon, How are you doing?
becomes---> Hello; Damon; How; are; you; doing.

**Parsing:** It is used to know the parts of speech

**Stemming:** It reduces the stress by removing the extra letters from the root word.
Ex: likely, liked, liking etc all will be treated as like.

**lemmatisation**: It is similar to stemming, but it gives us the accurate result.
Ex: chocolate, chocolatey, choco become choco
